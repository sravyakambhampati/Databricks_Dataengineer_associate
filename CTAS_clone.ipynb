* CTAS
* Constraints: Not null, Check

```
%sql 
create table example
(id Int, name String);
insert into example
values (1,'sravya');
```

CTAS Syntax

* Create table table1
AS select * from table2;

```
%sql
create table example2 as select * from example;
```

```
%sql 
select * from example2
```

* CTAS automatically infers schema from results of subquery and do not support manual schema declaration
* CTAS Additional Options Syntax


Create table name
comment "PII"
Partitioned by (column names)
Location "dbfs:/mnt/demo/"
AS select * from example;

```
%sql 
insert into example
values (2,'sri'),
(3,'vinnu'),
(4,'mommy');
```

```
%sql 
create table example3
comment "this is example 3"
partitioned by (id)
location "dbfs:/mnt/demo/example3"
as select id,name as emp_name from example;
```

```
%fs ls 'dbfs:/mnt/demo/example3'
```

#Constraints

```
%sql
Alter table example3 add constraint checkconstraint check(id<10); 
```

```
%sql
insert into example3
values (8,null);
```

```
%sql
Alter table example3 alter column emp_name set not null;
```

```
%sql
select * from example3;
```

```
%sql
create table example4
deep clone example3;
```

```
%sql
create table example_shallow
shallow clone example3;
```

Deep clone copies data and metadata from source to target where as shallow clone just copies transaction logs(creating copy of a table)

Deep clone occurs incrementally. So executing deep clone command again can synchronize changes
