```
dbs_tables= spark.sql("show tables in default like 'xxx'")

for i in spark.sql("show databases").collect():
    tables= spark.sql(f"show tables in {i.databaseName}")
    if tables.count()==0:
        tables=spark.sql(f"select '{i.databaseName}',null,null")
    dbs_tables=dbs_tables.union(tables)
dbs_tables.show()
```
