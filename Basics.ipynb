#What is Databricks?
Databricks is a multi cloud lakehouse platform based on Apache Spark. Lakehouse is unified analytics platform that combine features of datalake like openness, flexibility and ML  and datawarehouses features like governance,reliability and performance

#What is Databricks File sytem (DBFS)?
DBFS is a distributed file system mounted on Databricks workspaces and available in all databricks clusters. It is an abstraction layer on top of scalable storages like S3 providing us Unix-like file system features

* All purpose clusters: These can be shared between users and can be manually terminate and restart them
* Job clusters: These are ephimeral clusters. Databricks job schedular creates a cluster when you run a job on a new job cluster and terminates when the job is complete
