Data lake architecture lacks schema that is present in the datawarehouse

Delta lake is a component which is deployed on databricks clusters as part of databricks runtime.
Delta lake is a open source technology. delta lake is storage framework/layer and it enables building lakehouse architecture.
If we are creating a delta lake table, it gets stored in one or multiple files in parquet format. Along with these files, we will have transaction log also called as delta log. This log is ordered records of transactions performed on the table since its creation. It serves as single source of Truth, so everytime you query the table Spark reads this transaction log and retrieves the most recent version of the data
The logs are saved in json format.This json file consists of all the commited transactions or operations performed like insert, update, delete and predicates or filters used.

Delta lake guarantees that you will always get most recent version of data. Your will never have a deadlock state or conflicts with any ongoing operation on the table.
